```python
import pandas as pd
```

```python
## IMPORT DATA AS DATAFRAME
file = "C:\\Users\\Ly\\Desktop\\GitLabDemo\\MINSTP201808DATA2\\04-Pandas\\Homework\\HeroesOfPymoli\\Resources\\purchase_data.csv"
file_df = pd.read_csv(file)
file_df.head()
```

```python
## PLAYERS COUNT
players = file_df["SN"].unique()
total_players = len(players)
totalplayers = pd.DataFrame({
    "Total Players":[total_players]
})
totalplayers
```

```python
## PURCHASING ANALYSIS (TOTAL)
items = file_df["Item ID"].unique()
unique_items =len(items)
i_name = file_df["Item Name"].unique()

avg_price = file_df["Price"].mean()

total_purchases = file_df["Purchase ID"].count()

total_rev = file_df["Price"].sum()

purchasing_df = pd.DataFrame({
    "Number of Unique Items":[unique_items],
    "Average Price": [avg_price],
    "Number of Purchases": [total_purchases],
    "Total Revenue": [total_rev]
})
purchasing_df["Average Price"] = purchasing_df["Average Price"].map("${:.2f}".format)
purchasing_df["Total Revenue"] = purchasing_df["Total Revenue"].map("${:.2f}".format)
purchasing_df
```

```python
##GENDER DEMOGRAPHICS
males = file_df.loc[file_df["Gender"]== "Male",:]
females = file_df.loc[file_df["Gender"]== "Female",:]
other = file_df.loc[file_df["Gender"]== "Other / Non-Disclosed",:]

males_total = males["Gender"].count()
females_total = females["Gender"].count()
other_total = other["Gender"].count()

pct_males = (males_total/total_players)*100
pct_females = (females_total/total_players)*100
pct_other = (other_total/total_players)*100
gender_demo = pd.DataFrame({
    "Gender":["Male", "Female","Other/Non-Disclosed"],
    "Percentage of Players":[pct_males,pct_females,pct_other],
    "Total Count": [males_total,females_total,other_total]
})
gender_demo = gender_demo.set_index("Gender")
gender_demo["Percentage of Players"] = gender_demo["Percentage of Players"].map("{:.2f}".format)
gender_demo
```

```python
##PURCHASING ANALYSIS(GENDER)
avg_purchase_females = females["Price"].mean()
avg_purchase_males = males["Price"].mean()
avg_purchase_other = other["Price"].mean()

total_purchase_females = females["Price"].sum()
total_purchase_males = males["Price"].sum()
total_purchase_other = other["Price"].sum()

avg_total_females = total_purchase_females/females_total
avg_total_males = total_purchase_males/males_total
avg_total_other = total_purchase_other/other_total
```

```python
purchase_analysis = pd.DataFrame({
    "Gender":["Male", "Female","Other/Non-Disclosed"],
    "Purchase Count":[females_total,males_total,other_total],
    "Average Purchase Price": [avg_purchase_females,avg_purchase_males,avg_purchase_other],
    "Total Purchase Value": [total_purchase_females,total_purchase_males,total_purchase_other],
    "Avg Purchase Total per Person": [avg_total_females,avg_total_males,avg_total_other]
})
purchase_analysis = purchase_analysis.set_index("Gender")
purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis["Avg Purchase Total per Person"] = purchase_analysis["Avg Purchase Total per Person"].map("${:.2f}".format)
purchase_analysis
```

```python
##AGE DEMOGRAPHICS
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age = file_df["Age"]
price = file_df["Price"]
bins_df = pd.DataFrame({"Total Count": age,"Price":price})

bins_df["Age Bin"] = pd.cut(age, age_bins, labels=group_names)
age_df = bins_df.groupby("Age Bin").count()
age_df["Percentage of Players"] = age_df["Total Count"]/total_players*100
age_df = age_df[["Percentage of Players", "Total Count"]]
age_df["Percentage of Players"] = age_df["Percentage of Players"].map("{:.2f}%".format)
age_df
```

```python
##PURCHASING ANALYSIS (AGE)
price_df = bins_df.groupby("Age Bin").mean()
total_px = bins_df.groupby("Age Bin").sum()
pur_df = pd.DataFrame({"Total Count":age_df["Total Count"]})
pur_df["Average Purchase Price"] = price_df["Price"]
pur_df["Total Purchase Value"] = total_px["Price"]
pur_df["Average Purchase Total per Person"] = pur_df["Total Purchase Value"]/pur_df["Total Count"]
pur_df["Average Purchase Price"] = pur_df["Average Purchase Price"].map("${:.2f}".format)
pur_df["Total Purchase Value"] = pur_df["Total Purchase Value"].map("${:.2f}".format)
pur_df["Average Purchase Total per Person"] = pur_df["Average Purchase Total per Person"].map("${:.2f}".format)
pur_df
```

```python
##TOP SPENDERS
unique_players = pd.DataFrame({"SN": file_df["SN"], "Purchase Count": file_df["Purchase ID"]})
unique_players = unique_players.groupby("SN").count()
a = pd.DataFrame({"SN": file_df["SN"],"Average Purchase Price":file_df["Price"]})
a = a.groupby("SN").mean()
unique_players["Average Purchase Price"] = a["Average Purchase Price"]
unique_players["Total Purchase Value"] = unique_players["Purchase Count"]*unique_players["Average Purchase Price"]

unique_players = unique_players.sort_values("Total Purchase Value",ascending=False)
unique_players["Average Purchase Price"] = unique_players["Average Purchase Price"].map("${:.2f}".format)
unique_players["Total Purchase Value"] = unique_players["Total Purchase Value"].map("${:.2f}".format)

unique_players.head()
```

```python
## MOST POPULAR ITEMS
unique_items = pd.DataFrame({"Item ID": file_df["Item ID"],"Item Name": file_df["Item Name"],"Purchase Count":file_df["Price"]})
items_list = unique_items.groupby(['Item ID','Item Name']).count()
items_list["Item Price"] = unique_items.groupby(['Item ID','Item Name'])['Purchase Count'].mean()
items_list["Total Purchase Value"]=items_list["Purchase Count"]*items_list["Item Price"]
pop_item = items_list.sort_values("Purchase Count",ascending = False)
pop_item["Item Price"] = pop_item["Item Price"].map("${:.2f}".format)
pop_item["Total Purchase Value"] = pop_item["Total Purchase Value"].map("${:.2f}".format)
pop_item.head()
```

```python
## MOST PROFITABLE ITEMS
prof_item = items_list.sort_values("Total Purchase Value",ascending = False)
prof_item["Item Price"] = prof_item["Item Price"].map("${:.2f}".format)
prof_item["Total Purchase Value"] = prof_item["Total Purchase Value"].map("${:.2f}".format)
prof_item.head()
```
